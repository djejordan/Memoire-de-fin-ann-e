```python
# Suppression des variables obsolètes

df4 = df1.drop(columns = ["Team", "Games", "Season", "City"])
df4
```

```python
# Analyse statistique univariée du jeu de donnée

df4.describe()
```

```python
# Suppresion des doublons(s'ils existent)

df4.drop_duplicates(inplace = True)
```

```python
# Lister les valeurs manquantes
df4.isna().sum()
```

```python
# Remplacer les valeurs manquantes de Medal par Chocolate
df4["Medal"] = df4["Medal"].fillna("Chocolate")
```

```python
# Supprimer le reste des valeurs manquantes

df4.dropna(inplace = True)
```

```python
# Ré-indexons les lignes

df4.reset_index(inplace = True)
```

```python
# Repertorié toutes les discipline afin de supprimer celles qui ne sont plus d'actualités
df4["Sport"].unique()
```

```python
# Supprimer les athlets (Tug-Of-War,Art Competitions,Lacrosse)
df4 = df4.drop(df4[df4['Sport'] == 'Tug-Of-War'].index)
df4 = df4.drop(df4[df4['Sport'] == 'Art Competitions'].index)
df4 = df4.drop(df4[df4['Sport'] == 'Lacrosse'].index)
```

```python
df5 = df4.drop(columns = ["Sport"])
```

```python
df5.info()
```

+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

```python
df5
```

```python
df6 = df5.sort_values(["ID","Year"],axis=0)
```

```python
# Créer la colonne number_of_participation
df6["Number_of_participation"] = 1
```

```python
df6
```

```python
df6.describe()
```

```python
# Suppression des colonnes jugées inutiles

df6.drop(columns = ["index", "ID", "Name", "Number_of_participation"], inplace = True)
```

```python
# Sauvegarde du fichier
df6.to_csv("datafile_clean.csv")
```

```python
# créons une liste des identifiants
list_ID = df6["ID"].tolist()

# supprimons les doublons de cette liste
list_ID = list(set(list_ID))
len(list_ID)
```

```python
for year in range(1932, 2017, 4):
    for matricule in list_ID:
        a = 1
        for i in range(len(df6)) :
            if (df6.loc[i, "ID"] == matricule) and (df6.loc[i, "Year"] == year):
                df6.loc[i, "Number_of_participation"] = a
            else :
                continue
    a = a + 1
```
