```python
import pandas as pd
```

```python
# importation du fichier

df = pd.read_csv("datafile_clean.csv", index_col = [0])
df
```

```python
# Samplons le jeu de données 

sampled_df = df.sample(frac = 0.05)
```

```python
sampled_df
```

## Transformation des variables qualitatives en variables quantitatives

```python
# Lister les colonnes qui sont dites qualitatives
list_cat = [j for j in df.columns if df.dtypes[j] == "object"]
list_cat
```

```python
#df1 = df.copy()
```

```python
df_dummies = pd.get_dummies(
    sampled_df, columns = ['Sex', 'NOC', 'Event'], prefix="Column", dtype = float)
```

```python
df_dummies
```

## Separation du jeu de données (training set / test set)

```python
# Séparation du jeu de données en training set et test set
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df_dummies, test_size = 0.3, 
                                       random_state = 1)
```

```python
# Variable à prédire (Medal)

Y_train = train_set["Medal"]
Y_test = test_set["Medal"]
```

```python
# Variables prédictives

X_train = train_set.drop(columns = ["Medal"])
X_test = test_set.drop(columns = ["Medal"])
```

## Normalisation des données

```python
# Données d'entraînement

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X1 = scaler.transform(X_train)
```

```python
# Données de test

scaler_ = StandardScaler()
scaler_.fit(X_test)
X2 = scaler.transform(X_test)
```

## Predictions
